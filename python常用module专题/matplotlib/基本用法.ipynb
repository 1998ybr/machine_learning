**Step.1** 

$max\ \ U(X) $

$s.t.\ \sum_{1}^{n}P_iX_i=I$

**Step.2**

$L(X, \lambda)=U(X)+\lambda (\sum_1^n P_iX_i-I)$

$\frac{\partial L}{\partial X_i}=\frac{\partial U}{\partial X_i}+\lambda P_i=MU_i+\lambda P_i=0$

$\frac{\partial L}{\partial \lambda}=\sum_1^nP_iX_i-I=0$

根据第一个一阶条件可以得出均衡条件为：$\frac{MU_i}{MU_j}=\frac{P_i}{P_j}$，$i\ne j$

**Step.3**

当边际效用之比等于价格之比时，达到消费者均衡。
