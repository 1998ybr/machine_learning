# 基本用法

```python
#魔法函数：在jupyter可以不用plt.show()，直接显示图片
%matplotlib inline
#魔法函数：配合matplotlib插件ipympl，可以实现在jupyter的交互效果
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```



<center>一个matplotlib生成图片的各部分名称<\center>

```python
x = np.linspace(-1,1,50)#等差数列
y1 = 2*x+1
y2 = x**2
y3 = x**3
```

```python
plt.figure()
plt.plot(x,y1)
```

# figure——整个画布

```python
plt.figure()#创建画布，如果不创建，那么每次画图用的都是上一个figure
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y3)
```

```python
plt.plot(x,y1,color="red",linewidth=1,linestyle="-")#本图在上面一个figure里
```

# axis——坐标轴

```python
plt.figure()
plt.ylim((-1,2))# 设置y轴的范围
plt.xlim((-2,3))# 设置x轴的范围
plt.xlabel("描述x轴代表的实际意义",fontproperties = 'SimHei')#fontproperties参数是为了让matplotlib现实中文输出
plt.ylabel("描述y轴代表的实际意义",fontproperties = 'SimHei')
plt.plot(x,y3)
```

```python
plt.figure()
plt.ylim((-1,2))# 设置y轴的范围
plt.xlim((-2,3))# 设置x轴的范围
plt.xlabel("描述x轴代表的实际意义",fontproperties = 'SimHei')#fontproperties参数是为了让matplotlib现实中文输出
plt.ylabel("描述y轴代表的实际意义",fontproperties = 'SimHei')
plt.xticks(np.linspace(-2,3,3))# 设置x轴的刻度
plt.yticks([-1,0,1.5,2],
           [r"$very\ bad$",r"$bad \alpha$","just so so","good"]# 加上$$显示出来的就是Markdown在数学公式模式下文字的书写
           #如果使用特殊符号，比如\alpha，必须要要加r
          )# 设置y轴的刻度，使每个标签对应相应的数字刻度
plt.plot(x,y3)
```

```python
plt.figure()
plt.ylim((-1,2))# 设置y轴的范围
plt.xlim((-2,3))# 设置x轴的范围
plt.xlabel("描述x轴代表的实际意义",fontproperties = 'SimHei')#fontproperties参数是为了让matplotlib现实中文输出
plt.ylabel("描述y轴代表的实际意义",fontproperties = 'SimHei')
plt.xticks(np.linspace(-2,3,3))# 设置x轴的刻度
plt.yticks([-1,0,1.5,2],
           [r"$very\ bad$",r"$bad \alpha$","just so so","good"]# 加上$$显示出来的就是Markdown在数学公式模式下文字的书写
           #如果使用特殊符号，比如\alpha，必须要要加r
          )# 设置y轴的刻度，使每个标签对应相应的数字刻度
ax=plt.gca()#gca表示get current axis
ax.spines['right'].set_color("none")#将上“脊”和右“脊”设置为消失
ax.spines["top"].set_color("none")
plt.plot(x,y3)
```

```python
plt.figure()
plt.ylim((-1,2))# 设置y轴的范围
plt.xlim((-2,3))# 设置x轴的范围
plt.xlabel("描述x轴代表的实际意义",fontproperties = 'SimHei',loc="right")#fontproperties参数是为了让matplotlib现实中文输出
plt.ylabel("描述y轴代表的实际意义",fontproperties = 'SimHei',loc="top")
plt.xticks(np.linspace(-2,3,3))# 设置x轴的刻度
# plt.yticks([-1,0,1.5,2],
#            [r"$very\ bad$",r"$bad \alpha$","just so so","good"]# 加上$$显示出来的就是Markdown在数学公式模式下文字的书写
#            #如果使用特殊符号，比如\alpha，必须要要加r
#           )# 设置y轴的刻度，使每个标签对应相应的数字刻度
ax=plt.gca()#gca表示get current axis
ax.spines['right'].set_color("none")#将上“脊”和右“脊”设置为消失
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")#将下“脊”设置为x轴
ax.yaxis.set_ticks_position("left")#将左“脊”设置为y轴
ax.spines["bottom"].set_position(("data",0)) ##使用data的方法，把x轴放到y=0的位置上
ax.spines["left"].set_position(("axes",0.5)) ##使用axes的方法，把y轴定位到x轴的中心位置
plt.plot(x,y3)
```
