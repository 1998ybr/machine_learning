# 1 Python环境的简单介绍

Python是一种面向对象的高级编程语言，相对于C++和Java等语言，Python入门门槛低，让开发者能够用更少的代码实现想法；相应的，相对于C++和Java，Python也存在CPU利用效率低等问题，但是如果只是数据分析或非大型程序，Python的性能一般足够。

Python官网自带的IDE（集成开发环境，也可以简单地理解为Python的编辑器）即IDLE仅支持最基本的编辑功能和调试，因此使用Python还需要另外下载使用其他的IDE。

支持Python的IDE有很多，比较常见的有Pycharm、Spyder和Jupyter等。其中Pycharm功能相对强大，适合Python项目开发与团队合作，其自动补全功能和调试功能也非常好用，但是相应的使用起来也比较“笨重”；Spyder相当于Pycharm的简化版，使用较为轻便，自动补全功能和调试功能也差强人意。

由于笔者使用Python主要用于数据分析，因此使用的IDE为jupyter lab。jupyter lab的优势在于它的notebook文件支持Markdown书写，同时相对于jupyter notebook，它的插件也更加丰富强大，可以通过安装插件实现其本身不具有的代码补全与调试功能。相对于Pycharm，jupyter lab更加轻便，每次启动不需要太长的初始化时间，notebook界面给人的感觉也更清爽。

# 2 Python环境的搭建——基于Anaconda 

Anaconda是一个集成的安装管理程序，用于科学计算，致力于简化包和模块的管理与部署。简单来说，利用Anaconda之后可以省去很多诸如环境变量配置等麻烦，而且还可以搭建R的环境，让你实现左手Python右手R。

## 2.4 几个注意事项

1.　对于国内的大部分网络，Anaconda内下载的速度可能并不算快，网上一些帖子会教你改变默认镜像而是用国内高校的镜像，这确实会提高下载和安装速度。但是近些年国内部分知名高校如清华大学的镜像已经不能使用，后续同济等镜像能否维持也不得而知，因此在使用时需要注意。

2.　Anaconda所安装的Python可能并非官方发布的最新版本，比如2021年1月30日时，官方发布的Python最新版本为3.9，而Anaconda支持的为3.8。因此，如果你有了解版本的特殊需要，可以在Anaconda Prompt中输入
```python -V```进行查看。
